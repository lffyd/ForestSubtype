```python
#author：fyd
import numpy as np
import pandas as pd
import altair as alt
import sklearn
from sklearn.model_selection import train_test_split
from flag import Flags
from processData.processData import load_biology_data
from sklearn.ensemble import RandomForestClassifier as rf
from sklearn.metrics import accuracy_score
import altair_viewer
import tensorflow as tf
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier as knn
from sClass.sModel import randomForest
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.utils import shuffle
from sklearn import metrics
from sClass.sMethodAssess import purity_score
```

```python
print("开始加载数据...")
(X,y) = load_biology_data(Flags)
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=0
)
print("加载数据完毕...")
```

```python
print("查看X_train...")
X_train
```

```python
print("查看y_train...")
y_train
```

```python
print("开始测试随机森林预测分数：")
print("开始寻找随机森林最大深度...")
max_depth_best = []
for i in range(1,20):
    clf = rf(max_depth=i, random_state=0)
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    max_depth_best.append([i, accuracy_score(y_test, y_pred)])
max_depth = pd.DataFrame(max_depth_best, columns= ["max_depth", "Accuracy"])
max_depth_index = max_depth[max_depth.loc[:,"Accuracy"].max()==max_depth.loc[:,"Accuracy"]].index
max_depth_num = max_depth_index[0]+1
```

```python
print("查看max_depth图表")
max_depth_chart= alt.Chart(max_depth).mark_line().encode(
    alt.X("max_depth", scale= alt.Scale(zero=False)),
    alt.Y("Accuracy", scale= alt.Scale(zero=False)), 
    tooltip= ["max_depth", "Accuracy"]
).properties(
    title="Random Forest Acuracy vs Max Depth"
)
max_depth_chart
```

```python
print("寻找随机森林最大深度结束，最大深度为：",max_depth_num)
print("随机森林开始学习：")
model_scores =[]
rf_model = randomForest()
rf_model.fit(X_train, np.ravel(y_train))
```

```python
y_pred = rf_model.predict(X_test)
rd_score = accuracy_score(y_test, y_pred)
model_scores.append(rd_score)
print("预测正确率：",rd_score)
```

```python
NMI_score = metrics.cluster.normalized_mutual_info_score(y_test.iloc[:,0].tolist(), y_pred)
print("NMI：",NMI_score)
```

```python
#随机森林分类器混淆矩阵
print("随机森林分类器混淆矩阵：")
conf_mat = confusion_matrix(y_test, y_pred, labels= rf_model.classes_)
ConfusionMatrixDisplay(conf_mat, display_labels=rf_model.classes_).plot()
```

```python
print("特征重要度排行：")
sorted_idx = (-rf_model.feature_importances_).argsort()[:10]
feature_data = pd.DataFrame(
        {"Feature": X.columns[sorted_idx], "Importance": rf_model.feature_importances_[sorted_idx]}).sort_values(
        by="Importance", ascending=False)
feature_data
```

```python
alt.Chart(feature_data).mark_bar().encode(
    y= alt.Y("Feature", sort=alt.EncodingSortField(field="Importance", op="count", order='ascending'), title="Gene"),
    x=alt.X('Importance:Q'),
    color= alt.Color("Feature", title= "Gene", legend= None),
    tooltip = ["Feature", "Importance"]
)
```

```python
alt.Chart(feature_data).mark_bar().encode(
    y= alt.Y("Feature", sort=alt.EncodingSortField(field="Importance", op="count", order='ascending'), title="Gene"),
    x=alt.X('Importance:Q'),
    color= alt.Color("Feature", title= "Gene", legend= None),
    tooltip = ["Feature", "Importance"]
)
```

```python
#'''k近邻'''
print("随机森林测试结束：")
print("k近邻测试开始：")
neighbors = []

for i in range(1,20):
    kclass= knn(n_neighbors=i)
    kclass.fit(X_train, y_train)
    y_pred= kclass.predict(X_test)
    neighbors.append([i, accuracy_score(y_test, np.ravel(y_pred))])
neighbors = pd.DataFrame(neighbors, columns= ["n_neighbors", "Accuracy"])
max_Accuracy_num = neighbors.loc[:,"Accuracy"].max()
model_scores.append(max_Accuracy_num)
neighbors.head()
```

```python
alt.Chart(neighbors).mark_line().encode(
    alt.X("n_neighbors", scale= alt.Scale(zero=False)),
    alt.Y("Accuracy", scale= alt.Scale(zero=False)),
    tooltip=["Accuracy", "n_neighbors"]
).properties(
    title="Accuracy vs Number of Neighbors"
)
```

```python
kclass= knn(n_neighbors=3)

# Fitting Model
kclass.fit(X_train, y_train)

# Obtaining prediction
y_pred_knn= kclass.predict(X_test)

# Getting the Confusion Matrix
knn_conf_mat = confusion_matrix(y_test, y_pred_knn, labels= kclass.classes_)
ConfusionMatrixDisplay(knn_conf_mat, display_labels=kclass.classes_).plot()
```

```python
#'''支持向量机'''
print("k近邻测试结束")
print("支持向量机测试开始:")
svm_class = []
for i in range(1,10):
    clf = make_pipeline(StandardScaler(), SVC(C=i))
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    svm_class.append([i, accuracy_score(y_test, np.ravel(y_pred))])
svm_class = pd.DataFrame(svm_class, columns= ["C parameter", "Accuracy"])
max_Accuracy_num = svm_class.loc[:,"Accuracy"].max()
model_scores.append(max_Accuracy_num)
svm_class.head()
```

```python
alt.Chart(svm_class).mark_line().encode(
    alt.X("C parameter", scale= alt.Scale(zero=False)),
    alt.Y("Accuracy", scale= alt.Scale(zero=False))
).properties(
    title="ACCURACY VS C PARAMETER SVM"
)
```

```python
svm_model = SVC(C=8)

# Fitting Model
svm_model.fit(X_train, y_train)

# Obtaining prediction
y_pred_svm = svm_model.predict(X_test)

# Getting the Confusion Matrix
svm_conf_mat = confusion_matrix(y_test, y_pred_svm, labels= svm_model.classes_)
ConfusionMatrixDisplay(svm_conf_mat, display_labels=svm_model.classes_).plot()
```

```python
#'''逻辑回归模型'''
print("支持向量机测试结束")
print("逻辑回归开始：")
log_class = LogisticRegression(random_state=0, solver="saga",max_iter=200)
log_class.fit(X_train, y_train)
log_class_score = log_class.score(X_test, y_test)
model_scores.append(log_class_score)
log_class_score
```

```python
y_pred_log = log_class.predict(X_test)
log_conf_mat = confusion_matrix(y_test, y_pred_log, labels= log_class.classes_)
ConfusionMatrixDisplay(log_conf_mat, display_labels=log_class.classes_).plot()
```

```python
#'''多层感知器分类器'''
print("逻辑回归测试结束")
print("多层感知器分类器开始：")
clf = MLPClassifier(solver='adam', alpha=1e-5,hidden_layer_sizes=(1024,512), random_state=1,max_iter=1000) # 创建神经网络分类器对象
clf.fit(X_train, y_train)# 训练模型
clf_score = clf.score(X_test, y_test)#模型评分
model_scores.append(clf_score)
clf_score
```

```python
clf.fit(X_train, y_train)# 训练模型
y_pred_clf = clf.predict(X_test)
conf_mat_clf = confusion_matrix(y_test, y_pred_clf, labels= clf.classes_)
ConfusionMatrixDisplay(conf_mat_clf, display_labels=clf.classes_).plot()
```

```python
#'''组合分类器'''
print("多层感知器分类器测试结束")
print("组合分类器开始：")
rf_clf = rf(max_depth=14, random_state=0)
svm_clf = SVC(C=8)
log_clf = LogisticRegression(random_state=0, solver="saga", max_iter=200)
clf = MLPClassifier(solver='adam', alpha=1e-5,hidden_layer_sizes=(1024,512), random_state=1,max_iter=1000) # 创建神经网络分类器对象

ensembl_clf = VotingClassifier(
        estimators=[("rf", rf_clf), ("log_clf", log_clf)],
        voting="hard"
)
#
ensembl_clf.fit(X_train, y_train)

ensembl_clf_score = ensembl_clf.score(X_test, y_test)
model_scores.append(ensembl_clf_score)
ensembl_clf_score
```

```python
y_pred = ensembl_clf.predict(X_test)
conf_mat = confusion_matrix(y_test, y_pred, labels= ensembl_clf.classes_)
ConfusionMatrixDisplay(conf_mat, display_labels=ensembl_clf.classes_).plot()
```

```python
#提取随机森林TOP200特征
print("提取随机森林TOP200特征....")
merge_data_feature = pd.DataFrame(columns={"Feature":"","Importance":""},index=[0])
dd = []
for i in range(1,11):
    print("第",i,"次实验....")
    #随机打乱数据做10次实验取特征交集，最终的merge_Data就是选择出的TOP200特征
    X, y = shuffle(X, y)  # 第一重随机打乱
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=0,shuffle=True)  #第二重随机打乱
    
    rf_model = randomForest()
    rf_model.fit(X_train, y_train)

    sorted_idx = (-rf_model.feature_importances_).argsort()[:6666]
    feature_data = pd.DataFrame(
            {"Feature": X.columns[sorted_idx], "Importance": rf_model.feature_importances_[sorted_idx]}).sort_values(
            by="Importance", ascending=False)
    if merge_data_feature.empty:
        merge_data_feature = feature_data
    else:
        merge_data_feature = pd.merge(merge_data_feature.loc[:,"Feature"], feature_data.loc[:,"Feature"], on=['Feature'])
print("提取随机森林TOP200特征结束....")
merge_data_feature = merge_data_feature.loc[:200,"Feature"]
merge_data_feature
```

```python
#截取TOP200特征数据子集
data = X.loc[:, merge_data_feature.T.tolist()]
data
```
